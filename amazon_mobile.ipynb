```python
import pandas as pd
import html5lib
import requests
from bs4 import BeautifulSoup
import re
```

```python
Product = []
Price = []
Rating = []

def mobile_under_20000(url):
    req = requests.get(url)
    soup = BeautifulSoup(req.content, 'html.parser')
    
    product = soup.find_all('a',class_="a-link-normal a-text-normal")
    for i in product:
        Product.append(i.get_text().replace('\n','').replace('\n',''))
        
    price = soup.find_all('span',class_="a-price-whole")
    for i in price:
        Price.append(i.get_text().replace(',',''))
        
    rating = soup.find_all('div',class_='a-row a-size-small') 
    for i in rating:
        Rating.append(i.get_text().replace('\n','').split(' ')[0])
```

```python
mobile_under_20000("https://www.amazon.in/Smartphones-Under-%E2%82%B920-000-Basic-Mobiles/s?rh=n%3A1805560031%2Cp_36%3A-2000000")
```

```python
data = pd.DataFrame({})
data['Mobile'] = Product
data['Price'] = Price
data['Rating'] = Rating
data
```
