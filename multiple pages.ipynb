```python
Product = []
Price = []
Rating = []
next_urls = []
Img_url = []

import requests
import time
from bs4 import BeautifulSoup

BASE_URL = 'https://www.amazon.in'

def extract_data(soup):
    product = soup.find_all('a',class_="a-link-normal a-text-normal")
    for i in product:
        Product.append(i.get_text().replace('\n','').replace('\n',''))
       
    price = soup.find_all('span',class_="a-price-whole")
    for i in price:
        Price.append(i.get_text().replace(',',''))
       
    rating = soup.find_all('div',class_='a-row a-size-small')
    for i in rating:
        Rating.append(i.get_text().replace('\n','').split(' ')[0])
    
    imgurl = soup.find_all('div',class_='a-section aok-relative s-image-fixed-height')
    for i in imgurl:
        Img_url.append(i.get('img')['src'])
    
    
    
def mobile_under_20000(url):
    req = requests.get(url)
    if req.status_code == 200:
        soup = BeautifulSoup(req.content, 'html.parser')
        extract_data(soup)
        pagination_tag = soup.find_all('ul', class_='a-pagination')
        if any(pagination_tag):
            for each in pagination_tag:
                li_tags = each.find_all('li')
                for li_tag in li_tags:
                    a_tag = li_tag.find('a')
                    if a_tag and a_tag.get_text() != '1':
                        next_urls.append(BASE_URL + a_tag['href'])
    else:
        time.sleep(1)
        print("Trying Again..")
        mobile_under_20000(url)

def wrapper_func(url):
    next_urls.append(url)
    counter = 0
    while len(Product) < 500 and counter <= len(next_urls):
        print("Counter: ", counter)
        mobile_under_20000(next_urls[counter])
        counter += 1

wrapper_func("https://www.amazon.in/Smartphones-Under-%E2%82%B920-000-Basic-Mobiles/s?rh=n%3A1805560031%2Cp_36%3A-2000000")
```
