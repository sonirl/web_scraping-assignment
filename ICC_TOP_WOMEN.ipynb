```python
import pandas as pd
from bs4 import BeautifulSoup
import html5lib
import requests
import re
```

# TOP 10 WOMEN'S TEAM

```python
Team = []
Match = []
Points = []
Rating = []

def ICC_TOP_10(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    team = soup.find_all("span",class_="u-hide-phablet")
    for i in team:
        Team.append(i.get_text())
    match_point = soup.find_all("td", class_="table-body__cell u-center-text")
    for i in range(0,len(match_point),2):
        Match.append(match_point[i].get_text().replace(',',''))
        Points.append(match_point[i+1].get_text().replace(',',''))
    match_eng = soup.find("td",class_="rankings-block__banner--matches").get_text()
    Match.insert(0,match_eng)
    point_eng = soup.find('td',class_="rankings-block__banner--points").get_text().replace(',','')
    Points.insert(0,point_eng)
    
    #Rating
    rating_all = soup.find_all('td', class_="table-body__cell u-text-right rating")
    for i in rating_all:
        Rating.append(i.get_text().replace('\n',''))
    rating_eng = soup.find('td', class_="rankings-block__banner--rating u-text-right").get_text().replace('\n','').strip()
    Rating.insert(0,rating_eng)

ICC_TOP_10('https://www.icc-cricket.com/rankings/womens/team-rankings/odi')
```

```python
ICC_top_10_teams_wm = pd.DataFrame({})
ICC_top_10_teams_wm['Teams'] = Team[0:10]
ICC_top_10_teams_wm['Match'] = Match[0:10]
ICC_top_10_teams_wm['Points'] = Points[0:10]
ICC_top_10_teams_wm['Rating'] = Rating[0:10]

ICC_top_10_teams_wm
```

```python
ICC_top_10_teams_wm.to_csv('ICC_top_10_teams_wm.csv')
```

# TOP 10 BATWOMEN

```python
Player_Name = []
Player_Team = []
Player_Rating = []

def ICC_TOP_10_BATSMEN_WOMEN(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    player1 = soup.find("div",class_="rankings-block__banner--name-large").get_text()
    player_all = soup.find_all('td',class_="table-body__cell rankings-table__name name")
    for i in player_all:
        Player_Name.append(i.get_text().replace('\n',''))
    Player_Name.insert(0,player1)
    #Extracting team of players
    player1team = soup.find('div',class_='rankings-block__banner--nationality').get_text().replace('\n','')
    playerallteam = soup.find_all('span',class_='table-body__logo-text')
    for i in playerallteam:
        Player_Team.append(i.get_text())
    Player_Team.insert(0,player1team) 
    #Extracting Rating
    player1rating = soup.find('div',class_="rankings-block__banner--rating").get_text().replace('\n','')
    playerallrating = soup.find_all('td',class_="table-body__cell rating")
    for i in playerallrating:
        Player_Rating.append(i.get_text().replace('\n',''))
    Player_Rating.insert(0,player1rating)
    
ICC_TOP_10_BATSMEN_WOMEN('https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting')
```

```python
ICC_TOP_10_BATSMEN_WOMEN = pd.DataFrame({})
ICC_TOP_10_BATSMEN_WOMEN ['Name'] = Player_Name[0:10]
ICC_TOP_10_BATSMEN_WOMEN ['Team'] = Player_Team[0:10]
ICC_TOP_10_BATSMEN_WOMEN ['Rating'] = Player_Rating[0:10]

ICC_TOP_10_BATSMEN_WOMEN 
```

```python
ICC_TOP_10_BATSMEN_WOMEN.to_csv('ICC_TOP_10_BATSMEN_WOMEN.csv')
```
