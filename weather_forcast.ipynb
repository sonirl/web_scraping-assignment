```python
import pandas as pd
import html5lib
from bs4 import BeautifulSoup
import requests
```

```python
Period = []
Short_desc = []
Temp = []
Temp_word = []
Desc = []

def San_F_Weather(url):
    req = requests.get(url)
    soup = BeautifulSoup(req.content,'html.parser')

    period_name = soup.find_all('p',class_="period-name")
    for i in period_name:
        Period.append(i.get_text())
        
    short_desc = soup.find_all('p',class_='short-desc')
    for i in short_desc:
        Short_desc.append(i.get_text())
        
    temp = soup.find_all('p',class_='temp')
    for i in temp:
        Temp.append(i.get_text().split(':')[1].strip())
        Temp_word.append(i.get_text().split(':')[0].strip())
        
    desc = soup.find_all('div',class_='col-sm-10 forecast-text')
    for i in desc:
        Desc.append(i.get_text())
    
San_F_Weather('https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YBwwSOgzZPY')
```

```python
Weather = pd.DataFrame({})
Weather['Period'] = Period
Weather['Short_desc'] = Short_desc
Weather['Temp'] = Temp
Weather['Temp_word'] = Temp_word
Weather['Desc'] = Desc[0:9]

Weather
```

```python
Weather.to_csv('Weather.csv')
```
